```python
import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y,label = "seno0", color = "red")

plt.title("grafico de seno")
plt.xlabel("eje x")
plt.ylabel("eje y")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Preparar los Datos ---

# Elegimos un rango simple para X: 100 puntos entre -10 y 10
x = np.linspace(-10, 10, 100)

# Calculamos Y usando tu frmula exacta
y = (x**2) + (2*x) - 4

# --- 2. Dibujar (Mtodo "Global" plt) ---

# Dibujar la lnea
plt.plot(x, y, color='green', label='f(x) = x^2 + 2x - 4')

# --- 3. Aadir Contexto ---
plt.title('Grfico de f(x) = x^2 + 2x - 4')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Aadir lneas de ejes para referencia
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Activar la leyenda (para que se vea el 'label')
plt.legend()

# --- 4. Mostrar ---
plt.show()
```
