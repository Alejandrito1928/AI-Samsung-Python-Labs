```python
# 1. Definimos los conjuntos (sets)
A = {1, 2, 3}
B = {3, 4, 5}

print(f"Conjunto A: {A}")
print(f"Conjunto B: {B}")

# 2. Investigamos las operaciones

# Unión (Juntar todo, sin duplicados)
union = A | B
print(f"Unión (A | B):     {union}")

# Intersección (Solo lo que comparten)
interseccion = A & B
print(f"Intersección (A & B): {interseccion}")

# Diferencia (Lo que está en A, pero no en B)
diferencia = A - B
print(f"Diferencia (A - B):   {diferencia}")

# Diferencia Simétrica (Lo que no comparten)
diff_simetrica = A ^ B
print(f"Dif. Simétrica (A ^ B): {diff_simetrica}")
```

```python
from itertools import chain, combinations

def get_powerset(iterable):
    """
    Genera el conjunto potencia de un iterable.
    Usa el método 'itertools' que es el más eficiente en Python.
    """
    # Convertimos el iterable a una lista primero
    s = list(iterable)
    
    # Usamos chain.from_iterable() para encadenar los resultados
    # de 'combinations' para cada longitud 'r' desde 0 hasta len(s)
    return list(chain.from_iterable(
        combinations(s, r) for r in range(len(s) + 1)
    ))

# --- Prueba ---
mi_set = {'a', 'b', 'c'}
potencia = get_powerset(mi_set)

print(f"Conjunto original: {mi_set}")
print(f"Conjunto potencia (total {len(potencia)} elementos):")
print(potencia)
```

```python
def get_powerset_iterative(s):
    # Empezamos con el conjunto vacío
    powerset = [ [] ]
    
    for elemento in s:
        # Creamos copias nuevas de los subconjuntos existentes
        # añadiendo el nuevo elemento
        nuevos_subconjuntos = []
        for subconjunto in powerset:
            nuevos_subconjuntos.append(subconjunto + [elemento])
            
        powerset.extend(nuevos_subconjuntos)
        
    return powerset
mi_lista = ['a', 'b', 'c']
potencia_iter = get_powerset_iterative(mi_lista)

print(f"\n--- Método Iterativo ---")
print(f"Conjunto original: {mi_lista}")
print(f"Conjunto potencia (total {len(potencia_iter)} elementos):")
print(potencia_iter)
```

```python
from itertools import combinations

def power(a):
    powerset = set() 
    
    for i in range(len(a) + 1):
        # Ahora powerset es un set, y puede hacer la unión '|'
        # (Tu generador 's for s...' es innecesario, pero funciona)
        powerset = powerset | set(combinations(a, i))
        
    return powerset

# Prueba
print(power({1, 2, 3}))
```

```python
# Paso 1: Importar las librerías
# 'plt' es el alias estándar para matplotlib.pyplot
# 'np' es el alias estándar para numpy
import matplotlib.pyplot as plt
import numpy as np

# Paso 2: Crear los datos
# Usamos np.linspace() para crear 100 puntos en el eje X, desde -5 hasta 5
# Esto es exactamente lo que hace el PDF [cite: 2705]
x = np.linspace(-5, 5, 100)

# Aplicamos la función matemática (nuestro "modelo")
y = x**2

# Paso 3: Crear el gráfico (Plot)
# plt.plot() dibuja la línea
plt.plot(x, y)

# Paso 4: AÑADIR ETIQUETAS (El paso profesional)
# Un gráfico sin etiquetas es inútil. Acostúmbrate a hacer esto SIEMPRE.
plt.title("Gráfico de Y = X^2")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

# Añadir una cuadrícula para que sea más fácil de leer
plt.grid(True)

# Paso 5: Mostrar el gráfico
# En Jupyter, esto muestra el gráfico en la salida de la celda
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Definir el Eje X
#    (inicio=0, fin=10, paso=0.1)
x = np.arange(0, 10, 0.1)

# 2. Calcular Y1 (f(x) = x^2 + 2x - 1)
#    Esta es la forma "vectorizada" (la forma correcta en Numpy)
y1 = x**2 + 2*x - 1

# 3. Calcular Y2 (g(x) = e^(x/2))
#    Usamos np.exp() que funciona con todo el array 'x'
y2 = np.exp(x / 2)

# 4. Graficar
#    Añadimos 'label' para la leyenda
plt.plot(x, y1, "r--", label="f(x) = x^2 + 2x - 1")
plt.plot(x, y2, "b--", label="g(x) = e^(x/2)")

# 5. Añadir etiquetas y leyenda para claridad
plt.title("Gráfico de f(x) y g(x)")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid(True)
    
# 6. Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 2, 301)
y1 = 3**x
y2 = 3**(x - 1) + 2
graph1, = plt.plot(x, y1)
graph2, = plt.plot(x, y2)
plt.legend(handles=(graph1, graph2), labels=(r'$y=3^x$', r'$y=3^{x-1}+2$'))
plt.show()

x = np.linspace(-1, 2, 301)
y1 = 2**(-x)
y2 = -2**(-x - 2)
graph1, = plt.plot(x, y1)
graph2, = plt.plot(x, y2)
plt.legend(handles=(graph1, graph2), labels=(r'$y=2^{-x}$', r'$y=2^{-x-2}$'))
plt.show()
```
