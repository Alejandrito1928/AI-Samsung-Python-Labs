```python
import numpy as np
import matplotlib.pyplot as plt

def frontera_decision(x):
    return -x + 10

np.random.seed(50)
num_puntos = 50
x_data = np.random.uniform(2, 12, num_puntos)
y_data = np.random.uniform(2, 12, num_puntos)
#operacion de clasificacion
#callcular los valores de la frontera de decision
clase = y_data > frontera_decision(x_data) 

clase_colores = np.where(clase, 'red', 'blue')

plt.figure(figsize=(8, 8))

x_linea = np.linspace(2, 12, 10)
# Calcular los valores de y en la frontera de decisin
y_linea = frontera_decision(x_linea)
plt.plot(x_linea, y_linea, color='black', linestyle='--', linewidth=2, label=r'Frontera: $y = -x + 10$')

plt.scatter(x_data, y_data, c=clase_colores, s=80, alpha=0.7)

plt.title('Clasificacin Binaria por Funcin de Primer Grado', fontsize=16)
plt.xlabel('Eje X (Actividad del Usuario)', fontsize=12)
plt.ylabel('Eje Y (Nivel de Alerta de Logs)', fontsize=12)

plt.text(10, 3, 'Clase 0 (Normal)', color='blue', fontsize=14, weight='bold')
plt.text(3, 11, 'Clase 1 (Amenaza)', color='red', fontsize=14, weight='bold')

plt.xlim(2, 12)
plt.ylim(2, 12)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(loc='lower left', fontsize=11)

plt.show()

print("La Regla de Decisin (Frontera) es la funcin: y = -x + 10")
```
