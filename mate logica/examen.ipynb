```python
import pandas as pd
import numpy as np

def load_boston():
    data_url = "http://lib.stat.cmu.edu/datasets/boston"
    raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
    data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
    target = raw_df.values[1::2, 2]
    return {'data': data, 'target': target}


boston=load_boston()
X=boston[‘data’]
y=boston[‘target’]
```

```python
import numpy as np

a = np.array([4, 5, 2, 2])
b = np.array([4, 0, 2, 0])
c = np.array([2, 2, 0, 1])

print(np.linalg.norm(a - b))
print(np.linalg.norm(b - c))
print(np.linalg.norm(c - a))
```

```python
cos_ab = 1 - np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
cos_bc = 1 - np.dot(b, c) / (np.linalg.norm(b) * np.linalg.norm(c))
cos_ca = 1 - np.dot(c, a) / (np.linalg.norm(c) * np.linalg.norm(a))

print(cos_ab)
print(cos_bc)
print(cos_ca)
```

```python
import sympy

x = sympy.symbols('x')
f = x**3 - 1
print(sympy.diff(f))

x, k = sympy.symbols('x k')
f = sympy.log(x**2 - 3*k)
print(sympy.diff(f, x))

x, a, b = sympy.symbols('x a b')
f = sympy.exp(a * x**b)
print(sympy.simplify(sympy.diff(f, x)))
```

```python
import sympy

x, y = sympy.symbols('x y')
f = sympy.exp(x**2 + 2*y**2)

print(sympy.diff(f, x))
print(sympy.diff(f, y))
print(sympy.diff(f, x, x))
print(sympy.diff(f, x, y))
print(sympy.diff(f, y, x))
print(sympy.diff(f, y, y))
```
