```python
#graficaremos una derivada con import matplotlib y numpy
import numpy as np
import matplotlib.pyplot as plt
#definimos la funcion
def f(x):
    return x**3 - 3*x**2 + 4
#definimos la derivada de la funcion
def df(x):
    return 3*x**2 - 6*x
#definimos el rango de x
x = np.linspace(-10, 10, 100)
#calculamos los valores de la funcion y su derivada
#y = f(x)
dy = df(x)
#creamos la grafica
#plt.plot(x, y, label='f(x)')
plt.plot(x, dy, label="f'(x)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define la funcin f(x)
def function_limit(x):
    # La frmula que pediste
    return (x**2 - 1) / (x - 1)

x1 = np.linspace(-3, 0.999, 100)
x2 = np.linspace(1.001, 5, 100)
x_valores = np.concatenate([x1, x2])

y_valores = function_limit(x_valores)

plt.figure(figsize=(9, 6))

plt.plot(x_valores, y_valores, label=r'$f(x) = \frac{x^2 - 1}{x - 1}$', color='blue', linewidth=2)

limit_x, limit_y = 1, 2
plt.scatter(limit_x, limit_y, color='white', edgecolor='blue', s=150, zorder=5, label=r'Agujero ($x=1$)')
plt.scatter(limit_x, limit_y, color='green', marker='x', s=100, zorder=5) 

# f'(x) = 1
plt.hlines(1, xmin=-3, xmax=5, color='red', linestyle='--', label=r'$f^{\prime}(x) = 1$')


plt.title('Grfica de la Funcin del Lmite (Discontinuidad Removible)', fontsize=16)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)


plt.axhline(0, color='gray', linestyle=':', linewidth=0.8) 
plt.axvline(0, color='gray', linestyle=':', linewidth=0.8)  
plt.grid(True, linestyle=':', alpha=0.6)

plt.ylim(-2, 6)
plt.legend(fontsize=12)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_function(x):
    return (1 + 1/x)**x

x_valores = np.logspace(0, 4, 100) 

y_valores = euler_function(x_valores)

e_real = np.e 

plt.figure(figsize=(10, 6))

plt.plot(x_valores, y_valores, label=r'$y = \left(1 + \frac{1}{x}\right)^x$', color='blue', linewidth=2)

plt.axhline(e_real, color='red', linestyle='--', label=r'Valor Real de $e \approx 2.71828$')

plt.title('Convergencia a la Constante de Euler ($e$)', fontsize=16)
plt.xlabel(r'x (tende a $\infty$)', fontsize=12)
plt.ylabel('Valor de la Funcin', fontsize=12)

plt.xscale('log') 

plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(fontsize=12)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Definicin de las funciones ---
# Exponencial natural: f(x) = e^x
def exponential_function(x):
    return np.exp(x)

# Logaritmo natural: g(x) = ln(x)
def logarithm_function(x):
    # El dominio de ln(x) es x > 0
    return np.log(x)

# --- 2. Preparacin de los Datos ---
x_exp = np.linspace(-2, 3, 200) 
y_exp = exponential_function(x_exp)

# Rango para ln(x) (debe ser > 0.01)
x_log = np.linspace(0.01, 3, 200)
y_log = logarithm_function(x_log)

# Lnea de inversin y=x
x_inverse = np.linspace(-2, 3, 200)
y_inverse = x_inverse

# --- 3. Configuracin y Creacin de la Grfica ---
plt.figure(figsize=(8, 8))

# Graficar e^x (Curva Azul)
plt.plot(x_exp, y_exp, label=r'$y = e^x$', color='blue', linewidth=2)

# Graficar ln(x) (Curva Roja)
plt.plot(x_log, y_log, label=r'$y = \ln(x)$', color='red', linewidth=2)

# Graficar la lnea y=x (Eje de Simetra)
plt.plot(x_inverse, y_inverse, label=r'$y = x$ (Eje de Simetra)', color='gray', linestyle='--')

# Marcar puntos clave
plt.scatter([0], [1], color='blue', zorder=5) # e^x pasa por (0, 1)
plt.scatter([1], [0], color='red', zorder=5)  # ln(x) pasa por (1, 0)

# Etiquetas y Ttulo
plt.title('Relacin Inversa entre Exponencial y Logaritmo Natural', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)

plt.xlim(-2, 3)
plt.ylim(-2, 5)

plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(fontsize=12)

plt.show()
```
