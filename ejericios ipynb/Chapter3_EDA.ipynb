    matriz_A = np.array([[1, 2, 3], [4, 5, 6]]) \
    matriz_B = np.array([[7, 8, 9], [10, 11, 12]]) \
Realiza las siguientes operaciones usando np.append() y analiza los resultados:
- Concatenaci√≥n sin axis: Aplica np.append() sin especificar el par√°metro axis
- Concatenaci√≥n con axis=0: Usa np.append() con axis=0 para a√±adir filas
- Concatenaci√≥n con axis=1: Usa np.append() con axis=1 para a√±adir columnas

```python
import numpy as np
matriz_A = np.array([[1,2,3],[4,5,6]])
    
matriz_B = np.array([[7,8,9],[10,11,12]])

juntar = np.append(matriz_A, matriz_B, axis=0)
print(juntar)
juntar2 = np.append(matriz_A, matriz_B, axis=1)
print(juntar2)
```
