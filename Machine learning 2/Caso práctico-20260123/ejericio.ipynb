```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

data = pd.read_csv('kidney_disease.csv')
print(data.columns)
display(data.head())
data.info()
```

```python
if 'id' in data.columns:
    data.drop(['id'], axis=1, inplace=True)
data = data.select_dtypes(include=[np.number])
```

```python
# 1. Crear la copia con el nombre que pide la profesora
ckd_processed = data.copy()

# 2. Borrar la columna 'classification'
# Nota: "axis=1" significa borrar columna vertical
# "inplace=True" significa guardar el cambio ahÃ­ mismo
if 'classification' in ckd_processed.columns:
    ckd_processed.drop("classification", axis=1, inplace=True)

print("Columna 'classification' borrada.")
print("Dimensiones de ckd_processed:", ckd_processed.shape)
```

```python
from sklearn.preprocessing import StandardScaler

# 1. Crear el escalador
scaler = StandardScaler()

# 2. Ajustar y transformar los datos limpios (ckd_processed)
# Esto devuelve un array de numpy, no un dataframe
ckd_scaled_array = scaler.fit_transform(ckd_processed)

# 3. (Opcional) Convertirlo de nuevo a DataFrame para que se vea bonito
ckd_scaled = pd.DataFrame(ckd_scaled_array, columns=ckd_processed.columns)

print("Datos estandarizados (Media=0, Varianza=1).")
display(ckd_scaled.head())
```

```python
# Vamos a probar desde 1 hasta 10 clusters para ver cuÃ¡l es el mejor
inertias = []
rango_k = range(1, 11)

for k in rango_k:
    kmeans_temp = KMeans(n_clusters=k, random_state=42)
    kmeans_temp.fit(ckd_scaled)
    inertias.append(kmeans_temp.inertia_)

print("Inercias calculadas para 10 k's diferentes.")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(rango_k, inertias, marker='o', linestyle='--')
plt.title('MÃ©todo del Codo (Elbow Method)')
plt.xlabel('NÃºmero de Clusters (k)')
plt.ylabel('Inercia (Distancia intra-cluster)')
plt.xticks(rango_k)
plt.grid(True)
plt.show()
```
