```python
import nltk
import numpy as np
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import pairwise_distances
```

### 1. TF IDF representation:

#### 1.1. Create a TF IDF matrix: 

```python
my_docs = [
    "Cybersecurity is like protecting your own home but in the digital world where your computer is the house and the internet is a public street full of strangers. Imagine that your IP address is the postal address of your house and open ports are the doors and windows where information constantly enters and leaves. If you leave a window open without surveillance anyone can enter to snoop or steal your private data that is why we use tools like firewalls that work like private security guards watching 24 hours who enters and who leaves and blocking any suspect who is not on the guest list. Hackers are like modern thieves who do not need to break glass but look for rusty locks or unlocked windows known as vulnerabilities to sneak into your system without you noticing. Also there is malware which is like a biological virus or a spy that hides inside an apparently harmless email package that you open yourself confidently. For this reason encryption is fundamental in modern security because it converts your messages and passwords into secret mathematical codes that only you and the recipient with the correct key can read making it so that if someone intercepts the message on the way they only see meaningless scribbles. In short cybersecurity is not just having an antivirus installed it is a constant mindset of closing digital doors well and not trusting strangers who offer suspicious gifts on the network."
]
```

```python
# A very simple pre-processing.
my_docs = [x.lower() for x in my_docs]
```

TfidfVectorizer() arguments: <br>
- *max_features* : maximum number of features (distict words). <br>
- *min_df* : The minimum DF. Integer value means count and real number (0~1) means proportion. <br> 
- *max_df* : The maximum DF. Integer value means count and real number (0~1) means proportion. Helps to filter out the stop words. <br> 

```python
vectorizer = TfidfVectorizer(max_features = 20, min_df = 1, max_df = 3, stop_words = stopwords.words('english'))
X = vectorizer.fit_transform(my_docs).toarray()               # Conversion to a dense matrix with toarray().
```

```python
# Size of the X matrix (m x n).
X.shape
```

```python
# Output the features.
print(vectorizer.get_feature_names_out())
```

```python
# Output a vector corresponding to a document.
print(X[0])
```

#### 1.2. Calculate the cosine similarity:

```python
# The cosine similarity matrix.
1 - np.round(pairwise_distances(X, metric="cosine"),3)
```

```python
# Cosine similarity between two documents by linear algebra.
np.dot(X[0], X[0].T)
```
