```python
text = "Hola, hola, hola, Hola"

patron = r'(?:Hola)'
coincidencias = re.findall(patron, text)

print(coincidencias)
```

```python
text = "Hola, hola, hola, Hola"
patron = r'(?:Hola)'
coincidencias = re.findall(patron, text, re.IGNORECASE)

print(coincidencias)
```

```python
texto = "Mi número de teléfono es 123-456-7890."
patron = r'Mi número de teléfono es (?P<numero>\d{3}-\d{3}-\d{4}).'
coincidencia = re.search(patron, texto)
```

```python
coincidencia
```

```python
if coincidencia:
    numero_telefono = coincidencia.group('numero')
    print("Número de teléfono:", numero_telefono)
```

```python
texto = "Hola, mundo. Hola regex."

patron = r'Hola(?=, )'
coincidencias = re.findall(patron, texto)

print(coincidencias)
```

```python
texto = "Hola, mundo. Hola regex."

patron = r'Hola(?=, )'
coincidencias = re.findall(patron, texto)

print(coincidencias)
```

```python
m = re.search(r"Hello , my name is (\w+) (?=.*)", "Hello , my name is Antonio Banderas")
print(m)
```

```python
texto = "Hello, my name is John Brown."
patron = r'John(?!Brown)'
coincidencias = re.findall(patron, texto)

print(coincidencias)
```

```python
texto = "Hello, my name is John Brown."
patron = r'John(?!\s+Brown)'
coincidencias = re.findall(patron, texto)

print(coincidencias)
```
