Ejercicio 1 – Buscar una palabra (case-insensitive) \
Comprueba si una cadena contiene la palabra "python" sin importar mayúsculas o minúsculas.

```python
#Ejercicio 1 – Buscar una palabra (case-insensitive) \
#Comprueba si una cadena contiene la palabra "python" sin importar mayúsculas o minúsculas.
def contains_python(text):
    pattern = re.compile(r'python', re.IGNORECASE)
    return bool(pattern.search(text))
print(contains_python("I love Python programming."))  # True
```

Ejercicio 2 – Validar número entero \
Determina si una cadena representa un entero válido (positivo o negativo).

```python
#deermina si una cadena representa un entero valido 
def is_valid_integer(text):
    pattern = r'^[+-]?\d+$'
    return bool(re.match(pattern, text))
print(is_valid_integer("123"))      # True
```

Ejercicio 3 – Extraer números de un texto \
Extrae todas las secuencias de dígitos de una cadena. Ejemplo: Tengo 2 gatos y 15 peces

```python
#ejericio 3 – Extraer numeros de un texto 
#extrae todas las secunecias de lo gigitos de una cadena 
#EJEMPLO tengo 2 gatos y 15 peces
def extract_numbers(text):
    pattern = r'\d+'
    return re.findall(pattern, text)

print(extract_numbers("Tengo 2 gatos y 15 peces."))  
```

```python
#ejercio 4 validad email simple 
def is_valid_email(email):
    pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    return bool(re.match(pattern, email))
print(is_valid_email("alerex10@gmail.com"))
```

```python
#ejericio 5 Extraer palabras ignonrando puntuacion
def extract_words(text):
    pattern = r'\b\w+\b'
    return re.findall(pattern, text)
print(extract_words("Hello, world! This is a test."))
```

Ejercicio 6 – Extraer día, mes y año \
Extrae día, mes y año de una fecha DD/MM/YYYY.

```python
#ejericio 6  extraer dia mes y alo
def extract_date_components(date_str):
    pattern = r'(\d{2})/(\d{2})/(\d{4})'
    match = re.match(pattern, date_str)
    if match:
        day, month, year = match.groups()
        return day, month, year
    return None
print(extract_date_components("01/01/2005"))  # ('01', '01', '2005')
print(extract_date_components("2023-12-25"))  # None
```

Ejercicio 7 – Reordenar fechas con re.sub() \
Convierte fechas de YYYY-MM-DD a DD/MM/YYYY.

```python
import re

def reorder_date_pro(date_str):
    pattern_a = r'(\d{2})[-/](\d{2})[-/](\d{4})'
    pattern_b = r'(\d{4})[-/](\d{2})[-/](\d{2})'

    if re.match(pattern_a, date_str):
        return re.sub(pattern_a, r'\3-\2-\1', date_str)
    elif re.match(pattern_b, date_str):
        return re.sub(pattern_b, r'\1-\2-\3', date_str)
    
    return date_str

print(reorder_date_pro("25-12-2023"))
print(reorder_date_pro("2005/01/02"))
```

Ejercicio 8 – Detectar palabras duplicadas consecutivas  \
Detecta si hay palabras repetidas seguidas.

```python
#ejercioo 8 detectar palabras duplicadas y consecutivas 
def find_consecutive_duplicates(text):
    pattern = r'\b(\w+)\s+\1\b'
    return re.findall(pattern, text, re.IGNORECASE)
print(find_consecutive_duplicates("This is is a test test."))  # ['is', 'test']
```
