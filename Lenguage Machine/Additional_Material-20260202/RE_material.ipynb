INTRODUCCIÓN A EXPRESIONES REGULARES \
Una Expresión Regular (Regex) es una secuencia de caracteres que define un patrón de búsqueda. \
Es una técnica desarrollada en la ciencia computacional teórica y en la teoría de lenguajes formales. \
   Generalmente, las expresiones regulares se utilizan en: \
       -Motores de búsqueda \
       -Cuadros de diálogo de búsqueda y reemplazo en editores de texto \
       -Utilidades de procesamiento de texto como sed y AWK \
       -Análisis léxico

Las expresiones regulares (Regex) se definen basándose en el Lenguaje Regular (RL), que es independiente del Lenguaje de Programación (PL). \
Cada lenguaje de programación (PL) implementa la funcionalidad y capacidades del Lenguaje Regular (RL). \
    Ejemplo: la expresión regular seriali[sz] coincide con serialise y también coincide con serialize así que [sz] significa: coincidir con un carácter, ya sea s o z

An important concept in Regex is the pattern \
-The pattern specifies a “template” to match against a given string  \
Example: the pattern ".*456.*" \
matches the string "456" \
matches the string "123456" \
matches the string "123456789" \
The character "." means ‘match any character’ \
The character "*" means ‘match 0 or more repetitions’ \

Match object es un objeto de coincidencia, contiene información sobre la coincidencia encontrada entre la expresión regular y el texto analizado

```python
m = re.match(".*456.*" ,"456")
print(m)
```

```python
m = re.match(".*456.*" ,"123456")
print(m)
```

```python
m = re.match ( ".*456.*" ,"123456789" )
print(m)
```

En el ejemplo:
m = re.match(".*456.*", "123456789")
print(m)  # Output: <re.Match object; span=(0, 9), match='123456789'>

El patrón .*456.* está buscando cualquier cadena que contenga "456", pero el .* antes y después del "456" permite que coincidan con cualquier número de caracteres (incluyendo cero) antes y después del "456". En este caso, la cadena completa "123456789" coincide con el patrón, y por eso el match object devuelve la coincidencia completa, es decir, "123456789".
¿Cómo accedemos al substring "456"? Para acceder específicamente al substring "456", necesitas modificar el patrón para capturar esa parte de la cadena usando grupos. Los grupos se definen con paréntesis (). De esta forma, el grupo que contiene "456" será capturado y podemos acceder a él de manera separada.

```python
m = re.match(".*(456).*", "123456789")

if m:
    print(m.group()) # group() (sin argumentos) devuelve la coincidencia completa
```

```python
# Con el método group(1), se accede al primer grupo capturador, que es la subcadena "456" en este caso.
if m:
    print(m.group(1))
```

```python
# Carácter de punto "." - En una expresión regular coincide con cualquier carácter excepto el salto de línea.

resultado = re.search(r'Co.k.e', 'Cookie') # "." puede coincidir con cualquier carácter en esas posiciones
print(resultado.group())
```

```python
# "\w" (minúscula) coincide con cualquier letra, dígito o guion bajo (_) - Se refiere a cualquier carácter que forme parte de una palabra.

resultado = re.search(r'Co\wk\we', 'Cookie')
print(resultado.group())

#En este caso "\w" coincide con las letras "o" e "i" en "Cookie", devolviendo la coincidencia completa de la palabra.
```

```python
# "\W" (mayúscula) es lo opuesto de \w. 
# Mientras que \w coincide con cualquier letra, dígito o guion bajo
# "\W" coincide con cualquier carácter que no sea una letra, un dígito o un guion bajo. Coincide con cualquier otro carácter, como símbolos, espacios o signos de puntuación.

resultado = re.search(r'C\Wke', 'C@ke')
print(resultado.group())

# el @ es un carácter especial que no es alfanumérico (coincide con \W). Devuelve la coincidencia completa de la palabra.
```

```python
# "\s" (minúscula) coincide con cualquier carácter de espacio en blanco. Esto incluye el espacio regular ( ), saltos de línea (\n), tabulaciones (\t) y retornos de carro (\r).

resultado = re.search(r'Eat\scake', 'Eat cake')
print(resultado.group()) 

# \s coincide con el espacio regular entre las palabras "Eat" y "cake".
```

```python
# "\S" es lo opuesto de \s. 
# Mientras que \s coincide con cualquier carácter de espacio en blanco, 
# \S coincide con cualquier carácter que no sea un espacio en blanco.

resultado = re.search(r'Cook\Se', 'Cookie')
print(resultado.group())

#En este caso, encuentra la letra "i" en la palabra "Cookie", devolviendo la coincidencia completa de la palabra.
```

```python
#Ejemplos adicionales:
#El patrón r'Cook\Se' podría coincidir con otras palabras que tengan cualquier carácter que no sea un espacio en blanco entre "Cook" y "e"
#como "Cook9e", "Cook#e", o incluso "Cook.e".
m = re.search(r'Cook\Se', 'Cook9e').group()  # Coincidiría con 'Cook9e'
n = re.search(r'Cook\Se', 'Cook.e').group()  # Coincidiría con 'Cook.e'
print(m)
print(n)
```

```python
# "\t" coincide con un carácter de tabulación. 
# La tabulación es un tipo específico de espacio en blanco que se usa a menudo para alinear texto en documentos o para crear espacios más grandes entre elementos.

resultado = re.search(r'Eat\tcake', 'Eat\tcake') #Texto = Eat   cake
print(resultado.group())  # Output: 'Eat\tcake'

# \t representa un carácter de tabulación, que es un tipo de espacio en blanco más amplio que un espacio regular. 
# La tabulación se usa a menudo para crear indentaciones o para alinear texto en columnas.
```

```python
# Ejemplo adicional:
# La coincidencia se activará solo si hay un carácter de tabulación entre "Eat" y "cake". Si se intenta hacer coincidir con un espacio regular, la búsqueda falla

re.search(r'Eat\tcake', 'Eat cake')  # No coincide, ya que hay un espacio en lugar de una tabulación.
```

Descripción de los metacaracteres: \
"\n": Coincide con un salto de línea. Este metacaracter se utiliza para indicar el final de una línea y el comienzo de otra. \
"\r": Coincide con un retorno de carro. \
"\d": Coincide con cualquier dígito decimal (0-9). Este metacaracter es útil para buscar números dentro de cadenas de texto.

```python
resultado = re.search(r'c\d\dkie', 'c00kie')
print(resultado.group())  # Output: 'c00kie'

#Análisis del Patrón: r'c\d\dkie'
# c: Coincide exactamente con la letra "c".
# d: Coincide con un dígito decimal. En este caso, coincide con "0".
# \d: Coincide nuevamente con otro dígito decimal. En este caso, coincide también con "0".
# kie: Coincide exactamente con las letras "kie".
```

```python
# "^"" se conoce como caret. Este metacaracter se utiliza para indicar que la coincidencia debe encontrarse al inicio de la cadena.

resultado = re.search(r'^Eat', 'Eat cake')
print(resultado.group())

# El patrón r'^Eat' coincide con la cadena 'Eat cake', devolviendo la coincidencia completa de 'Eat'
```

```python
# "$" se usa para indicar que la coincidencia debe encontrarse al final de la cadena.

resultado = re.search(r'cake$', 'Eat cake')
print(resultado.group())  # Output: 'cake'

#El patrón r'cake$' coincide con la cadena 'Eat cake', devolviendo la coincidencia completa de 'cake'.
```

```python
# La sintaxis de corchetes [...] se emplea para indicar un conjunto de caracteres. 
# Esto permite especificar un conjunto de caracteres entre los cuales se puede encontrar una coincidencia.

m1 = re.search(r'reali[zs]e', 'realize')
print(m1.group())  # Output: 'realize'

m2 = re.search(r'reali[zs]e', 'realise')
print(m2.group())  # Output: 'realise'

# El patrón r'reali[zs]e' coincide con ambas cadenas 'realize' y 'realise', devolviendo las coincidencias completas.
```

```python
# La sintaxis de rango de caracteres [..-..] permite definir un rango de caracteres dentro de los corchetes. 
# Esto significa que la expresión regular puede coincidir con cualquier carácter que se encuentre dentro de ese rango.

resultado = re.search(r'Number: [0-6]', 'Number: 5')
print(resultado.group())

#El patrón r'Number: [0-6]' coincide con la cadena 'Number: 5', devolviendo la coincidencia completa.
```

```python
# La sintaxis [^...] se emplea para definir un conjunto de caracteres negativo. 
# Coincidirá con cualquier carácter que no esté incluido en el conjunto especificado.

resultado = re.search(r'Number: [^5]', 'Number: 0')
print(resultado.group())
```

```python
# El operador "|" permite crear una expresión regular que coincide con A o B.

resultado1 = re.match(r'user\d@[g|\w+]mail\.com', 'user1@gmail.com')
print(resultado1)

resultado2 = re.match(r'user\d@[g|\w]+mail\.com', 'user2@hotmail.com')
print(resultado2)
```

```python
# Es útil para definir opciones alternativas dentro de una expresión regular.

# Coincide con 'user1@gmail.com'
resultado1 = re.match(r'user\d@[g|w]+mail\.com', 'user1@gmail.com')
print(resultado1)

# Coincide con 'user2@hotmail.com'
resultado2 = re.match(r'user\d@[g|w]+mail\.com', 'user2@hotmail.com')
print(resultado2)
```

Descripción:
Patrón: r'user\d@[g|w]+mail\.com'
user: Coincide exactamente con la palabra "user".
\d: Coincide con un dígito decimal (0-9), permitiendo cualquier número después de "user".
@: Coincide exactamente con el carácter "@".
[g|w]+: Aquí se presenta el operador | dentro de los corchetes, lo que puede llevar a confusiones. En este contexto, [g|w]+ significa "una o más ocurrencias de 'g', '|', o 'w'". Por lo tanto, esto no funcionará como se espera para coincidir con un dominio. En lugar de usar | aquí, deberías especificar los caracteres deseados directamente, como [gw].
mail\.com: Coincide exactamente con "mail.com". Nota que se usa \. para escapar el punto, ya que en expresiones regulares un punto por sí solo coincide con cualquier carácter.

Cadena 1: 'user1@gmail.com'
    En esta cadena, coincide con la estructura esperada y, por lo tanto, se encuentra una coincidencia.
    Resultado: re.match() encuentra la coincidencia y devuelve un objeto de coincidencia.

Cadena 2: 'user2@hotmail.com'
    Aquí, el patrón intenta coincidir con "hotmail.com". Sin embargo, hotmail no coincide con el patrón @[g|w]+mail\.com, ya que g o w no son parte de "hotmail". Por lo tanto, no habrá coincidencia.
    Resultado: re.match() devuelve None.

```python
# El operador | se puede usar de la siguiente manera, sin corchetes
# Coincide con 'user1@gmail.com' o 'user1@hotmail.com'

resultado = re.match(r'user\d@(gmail\.com|hotmail\.com)', 'user1@gmail.com')
print(resultado.group())

resultado2 = re.match(r'user\d@(gmail\.com|hotmail\.com)', 'user1@hotmail.com')
print(resultado2.group())
```

```python
# "\A" coincide solo con el inicio de la cadena, independientemente de si la cadena se extiende a múltiples líneas o no.

resultado = re.search(r'\A[A-E]ookie', 'Cookie')
print(resultado.group())
```

```python
# "\b" se emplea para coincidir con los límites de una palabra, con la posición en la que una palabra comienza o termina

resultado1 = re.search(r'\b[A-E]ookie', 'Cookie')
print(resultado1.group())
```

Syntax of Regex patterns 
Character Description 
"\" Backslash. Used to escape the special meaning of "\" 
when used in a escape character. 
re.search ( r'Back\\stail' , 'Back\stail').group() 
' Back\\stail ' <- \ is not escaped 
re.search ( r'Back\stail' , 'Back tail').group () 
' Back tail ' <- "\s" is interpreted 

```python
# El carácter de escape \ (backslash) en expresiones regulares permite escapar el significado especial de otros metacaracteres. 
# Este carácter es crucial para asegurar que los siguientes caracteres se interpreten literalmente en lugar de como un símbolo especial en el contexto de las expresiones regulares.

resultado1 = re.search(r'Back\\stail', 'Back\stail')
print(resultado1.group())

# La expresión regular r'Back\\stail' está diseñada para buscar una cadena que contenga el texto literal Back\stail

resultado2 = re.search(r'Back\stail', 'Back tail')
print(resultado2.group())

# El patrón r'Back\stail' busca "Back" seguido de un espacio en blanco y luego "tail".
```

Caracteres de repetición: \
Los caracteres de repetición son metacaracteres utilizados en expresiones regulares que especifican cuántas veces un elemento (carácter, grupo o patrón) puede aparecer en una cadena de texto. Los principales caracteres de repetición son: \
"*" (Asterisco): Coincide con cero o más repeticiones del elemento que lo precede. \
    Ejemplo: a* coincidirá con '' (cadena vacía), a, aa, aaa, etc. \
"+" (Signo más): Coincide con uno o más repeticiones del elemento que lo precede. \
    Ejemplo: a+ coincidirá con a, aa, aaa, pero no con '' (cadena vacía). \
"?" (Signo de interrogación): Coincide con cero o una repetición del elemento que lo precede. \
    Ejemplo: a? coincidirá con '' (cadena vacía) y a. \
{n} (Llaves): Coincide con exactamente n repeticiones del elemento que lo precede. \
    Ejemplo: a{3} coincidirá con aaa. \
{n,} (Llaves con coma): Coincide con n o más repeticiones del elemento que lo precede. \
    Ejemplo: a{2,} coincidirá con aa, aaa, aaaa, etc. \
{n,m} (Llaves con dos valores): Coincide con entre n y m repeticiones del elemento que lo precede. \
    Ejemplo: a{2,4} coincidirá con aa, aaa y aaaa, pero no con a o aaaaa.

```python
# El metacaracter + se utiliza en expresiones regulares para coincidir con uno o más caracteres a la izquierda del +. 
# Esto significa que el patrón que precede al + debe aparecer al menos una vez para que se considere una coincidencia.

resultado = re.search(r'Co+kie', 'Cooookie')
print(resultado.group())
```

```python
# "*"" Se emplea en expresiones regulares para coincidir con cero o más caracteres a la izquierda del *. 
# Esto significa que el patrón que precede al * puede aparecer ninguna o varias veces para que haya una coincidencia.


resultado1 = re.search(r'Ca*o*kie', 'Caokie')
print(resultado1.group())

resultado2 = re.search(r'Ca*o*kie', 'Coooookie')
print(resultado2.group())
```

```python
# "?" verifica exactamente cero o una ocurrencia del carácter o patrón que lo precede. 
# El carácter debe estar presente como máximo una vez para que haya una coincidencia.

resultado1 = re.search(r'Colou?r', 'Color')
print(resultado1.group())

resultado2 = re.search(r'Colou?r', 'Colour')
print(resultado2.group())

resultado3 = re.search(r'Colou?r', 'Colouur')
print(resultado3)
```

```python
# {x} se emplea en expresiones regulares para indicar que el elemento que lo precede debe aparecer exactamente x veces. 
# Esto es útil cuando se desea especificar el número exacto de repeticiones de un carácter, grupo o patrón.

resultado1 = re.search(r'Value\s{1}\d+', 'Value 123')
print(resultado1.group())

resultado2 = re.search(r'Value\s{1}\d+', 'Value  345')
if resultado2:
    print(resultado2.group())
else:
    print("No match")

# El patrón r'Value\s{1}\d+' busca la cadena "Value" seguida de exactamente un espacio en blanco y luego uno o más dígitos
```

```python
# {x,} es un cuantificador en expresiones regulares que indica que el patrón que lo precede debe repetirse al menos x veces o más.
# x es un número entero que especifica el número mínimo de repeticiones requeridas
# No hay un límite máximo definido en este caso, lo que significa que puede haber una cantidad indefinida de repeticiones.

# Ejemplo 1
resultado1 = re.search(r'Value\s{1,}\d+', 'Value 123')
print(resultado1.group())

# Ejemplo 2
resultado2 = re.search(r'Value\s{1,}\d+', 'Value 345')
print(resultado2.group())

# El carácter {1,} en el patrón indica que debe haber al menos un espacio entre "Value" y los números. 
```

```python
# {x,y}: es otro cuantificador en expresiones regulares que indica que el patrón que lo precede debe repetirse al menos x veces pero no más de y veces.
# x es el número mínimo de repeticiones requeridas, y y es el número máximo de repeticiones permitidas.

resultado = re.search(r'\d{9,10}', '0987654321')
print(resultado.group())  # Output: '0987654321'

# {9,10} en el patrón asegura que se busquen secuencias de dígitos que tengan un total de 9 a 10 dígitos.
```

Coincidencia codiciosa (Greedy Matching): es un comportamiento de las expresiones regulares donde el motor de búsqueda intenta hacer coincidir la máxima cantidad posible de caracteres que cumplen con el patrón especificado.
Esto significa que, al encontrar una coincidencia, el motor seguirá buscando para extenderla tanto como sea posible antes de considerar otros patrones.

Coincidencia no codiciosa o perezosa (Non-Greedy Matching): busca la mínima cantidad posible de caracteres que cumplen con el patrón.
Se logra añadiendo un signo de interrogación ? después del cuantificador (por ejemplo, *?, +?, {x,y}?), lo que le indica al motor que debe intentar coincidir con la menor cantidad de caracteres necesarios.

```python
# Coincidencia Codiciosa (Greedy Match): intenta encontrar la coincidencia más larga posible.

heading = r'<h1>TITLE</h1>'
resultado = re.match(r'<.*>', heading)
print(resultado.group()) 

# Al usar .*, el motor de regex buscará la mayor cantidad posible de caracteres que coincidan con .* entre el primer < y el último >.
```

```python
# Coincidencia No Codiciosa (Non-Greedy Match): busca capturar la menor cantidad posible de texto que coincida con el patrón especificado.

heading = r'<h1>TITLE</h1>'
resultado_no_codicioso = re.match(r'<.*?>', heading)
print(resultado_no_codicioso.group()) 

# (.*?) busca la menor cantidad de texto posible entre los delimitadores < y >.
```

Regex Functions: Funciones de expresiones regulares
Las funciones de expresiones regulares son funciones proporcionadas por bibliotecas de programación (como el módulo re en Python) que permiten trabajar con patrones de texto a través de expresiones regulares. Estas funciones facilitan la búsqueda, coincidencia, manipulación y análisis de cadenas de texto utilizando patrones específicos.

```python
# search(...): escanea a través de la cadena/secuencia dada buscando la primera ubicación donde la expresión regular produce una coincidencia. 
# Devuelve un objeto de coincidencia correspondiente si se encuentra, de lo contrario devuelve None.

pattern = "cookie"
sequence = "Cake and cookie"
resultado = re.search(pattern, sequence)

# Si se encuentra una coincidencia, imprime el texto coincidente
if resultado:
    print(resultado.group())
    
# re.search busca la palabra "cookie" dentro de la cadena "Cake and cookie".
# Encuentra la coincidencia en la posición correspondiente y, al llamar a resultado.group(), devuelve la cadena coincidente: 'cookie'.
```

```python
# match(...): escanea a través de la cadena/secuencia dada buscando la primera ubicación donde la expresión regular produce una coincidencia. 
# Devuelve un objeto de coincidencia correspondiente si se encuentra, de lo contrario devuelve None.

pattern = "C"
sequence1 = "IceCream"
resultado1 = re.match(pattern, sequence1)


# Imprimimos el resultado
if resultado1 is None:
    print("No se encontró coincidencia.")
else:
    print(resultado1.group())
```

```python
sequence2 = "Cake"
resultado2 = re.match(pattern, sequence2)  # Esto devolverá un objeto de coincidencia

# Si se encuentra una coincidencia, imprime el texto coincidente
if resultado2:
    print(resultado2.group())
```

search() vs. match(): la función match() verifica si hay una coincidencia solo al comienzo de la cadena (por defecto). \
La función search() verifica si hay una coincidencia en cualquier parte de la cadena.

```python
# findall(...) busca todas las posibles coincidencias de un patrón específico en toda la secuencia (cadena de texto) proporcionada y las devuelve como una lista de cadenas.
# Cada cadena devuelta representa una coincidencia.

# Cadena que contiene direcciones de correo electrónico
email_address = "Please contact us at: support@datacamp.com, xyz@datacamp.com"

# Expresión regular que busca patrones de dirección de correo electrónico
pattern = r'[\w.-]+@[\w.-]+'

# Usamos findall para encontrar todas las coincidencias en la cadena
addresses = re.findall(pattern, email_address)

# Imprimimos cada dirección encontrada
for address in addresses:
    print(address)
    
# La función re.findall(...) busca en la cadena email_address y encuentra dos coincidencias: "support@datacamp.com" y "xyz@datacamp.com".
```

```python
# sub(...): se utiliza para sustituir partes de una cadena de texto que coinciden con un patrón específico por un nuevo valor (sustitución).
#  Devuelve una nueva cadena donde las ocurrencias del patrón se han reemplazado por el texto proporcionado como reemplazo (repl). 
# Si no se encuentra el patrón en la cadena, se devuelve la cadena original sin cambios.

# Cadena que contiene una dirección de correo electrónico
email_address = "Please contact us at: xyz@datacamp.com"

# Usamos sub para reemplazar el correo electrónico encontrado por uno nuevo
new_email_address = re.sub(
    r'([\w.-]+)@([\w.-]+)',  # Patrón para encontrar el correo electrónico
    r'support@datacamp.com', # Reemplazo
    email_address            # Cadena original
)

# Imprimimos la nueva dirección de correo electrónico
print(new_email_address)

# La función re.sub(...) busca en la cadena email_address y encuentra la coincidencia xyz@datacamp.com. Luego, reemplaza esa coincidencia con support@datacamp.com.
```

```python
# En este caso, reemplaza la primera coincidencia del patrón (de izquierda a derecha) que no se superpone, por el nuevo valor.

# Cadena que contiene varias direcciones de correo electrónico
email_address = "Please contact us at: xyz@old.com@old.com"

# Usamos sub para reemplazar el primer correo electrónico encontrado por uno nuevo
new_email_address = re.sub(
    r'([\w.-]+)@([\w.-]+)',  # Patrón para encontrar el correo electrónico
    r'support@new.com',      # Reemplazo
    email_address            # Cadena original
)

# Imprimimos la nueva dirección de correo electrónico
print(new_email_address)

# La función re.sub(...) busca en la cadena email_address y encuentra la primera coincidencia xyz@old.com. 
# Luego, reemplaza esa coincidencia con support@new.com. 
# Sin embargo, observa que hay otra dirección de correo electrónico (@old.com) después de la primera, pero no se reemplaza porque la función sub solo se ocupa de la primera ocurrencia que encuentra.
```

```python
# compile(...): se utiliza para compilar una expresión regular en un objeto de expresión regular.
# Esto es útil cuando la misma expresión regular se utilizará varias veces en el código, ya que compilarla una vez y reutilizar el objeto resultante puede ser más eficiente que volver a compilar la misma expresión cada vez que se usa.

# Compilamos la expresión regular para buscar "cookie"
pattern = re.compile(r"cookie")

# Cadena en la que buscamos
sequence = "Cake and cookie"

# Usamos el objeto compilado para buscar en la cadena
result = pattern.search(sequence).group()

# Imprimimos el resultado
print(result)
```

```python
#Esto es equivalente a:
re.search(pattern, sequence).group()
```

Los Name Groups son una característica de las expresiones regulares que te permite asignar nombres a grupos de captura dentro de una expresión regular. Esto facilita el acceso a las coincidencias específicas de esos grupos en lugar de referirse a ellos solo por su posición (índice) en la lista de grupos de coincidencia.

```python
# Grupos No Capturantes (?:...): Un grupo no capturante se define utilizando la sintaxis (?:...).
# Este tipo de grupo se utiliza para agrupar patrones sin capturarlos para su uso posterior.
# Es útil cuando se necesita aplicar cuantificadores a un grupo de expresiones, pero no se desea que ese grupo se devuelva como parte de los resultados de coincidencia.

m = re.search(r"Hello, (?:my name is|I am) (.*)", "Hello, my name is Slim Shady")
print(m) 

#La expresión regular busca una coincidencia que comience con "Hello," seguido de una de las dos frases: "my name is" o "I am". 
# Este grupo se define como no capturante usando (?:...).
# El .* después del grupo capturante (.*) captura cualquier texto que siga, que en este caso es "Slim Shady".
```

```python
#Resultado: Aunque el grupo no capturante no se devuelve como parte del resultado, se captura "Slim Shady" en el grupo 1

print(m.group(1)) 
```

```python
# Los grupos nombrados se definen utilizando la sintaxis (?P<tag>...). 
# Esto te permite asignar un nombre al grupo, lo que facilita la referencia a ese grupo en lugar de usar un índice numérico. 
# Los grupos nombrados son particularmente útiles en expresiones regulares complejas o cuando hay múltiples grupos, ya que puedes acceder a ellos de forma más intuitiva.

m = re.search(r"Hello, (?:my name is|I am) (?P<name>.*)", "Hello, my name is Slim Shady")
print(m.group("name"))

# El mismo patrón se utiliza, pero el grupo que captura "Slim Shady" se nombra como name utilizando (?P<name>...).
# Se puede acceder a la coincidencia utilizando m.group("name"), que devolverá "Slim Shady".
```

```python
# El método groupdict() devuelve un diccionario que contiene todos los grupos nombrados, lo que permite acceder fácilmente a ellos por su nombre. 
# En este caso, el diccionario resultante sería {'name': 'Slim Shady'}.
print(m.groupdict())
```

```python
# La afirmación de Anticipación (Lookahead Assertion) se utiliza para buscar una coincidencia que debe estar presente en la cadena, pero que no debe formar parte de la coincidencia final. 
# Es una forma de "mirar hacia adelante" en la cadena sin consumir los caracteres que se están buscando. Esto significa que puedes comprobar si un patrón sigue a otro, sin que el patrón de anticipación forme parte del resultado.

m = re.search(r"Hello, my name is (\w+)(?=.*)", "Hello, my name is Antonio Banderas")
print(m) 

# El (?=.*) al final es una afirmación de anticipación que indica que después del nombre debe haber cualquier cosa (el .* permite cualquier carácter, incluyendo espacios, letras, etc.), pero no consume esos caracteres. 
# Esto significa que "Antonio" es capturado por el grupo 1,
```

```python
# Cuando se imprime m, se obtiene el objeto de coincidencia completo que incluye "Hello, my name is Antonio". 
# Al acceder al grupo 1 con m.group(1), obtienes solamente "Antonio".

print(m.group(1))  # 'Antonio'
```

```python
# La afirmación de Anticipación Negativa (Negative Lookahead Assertion) se utiliza para asegurarse de que una cierta cadena no siga a un patrón dado. 
# Esto significa que el motor de expresiones regulares verificará si un patrón específico no está presente después del texto que estás buscando. 
# Si el patrón no está presente, la coincidencia será válida. Este tipo de afirmación es útil cuando deseas excluir ciertos patrones de tus resultados.

m = re.search(r"Hello, my name is (\w+)(?!.+)", "Hello, my name is Antonio")
print(m)  # <re.Match object; span=(0, 25), match='Hello, my name is Antonio'>

#La afirmación de anticipación negativa (?!.+) verifica que después del nombre no haya más caracteres. Asegura que no haya más texto después de "Antonio".
```

```python
# Al acceder al grupo 1 con m.group(1), obtienes solamente "Antonio".

print(m.group(1))
```

```python
##Las expresiones negativas junto con la laxitud (laziness) en expresiones regulares se refiere a la forma en que un patrón puede coincidir con la menor cantidad de texto posible. 
# Cuando utilizas afirmaciones de anticipación negativa en combinación con otros patrones, puedes lograr que tu expresión regular sea más específica y menos propensa a consumir más texto del que realmente necesitas.

m = re.search(r"<title>((((?!<).)*))</title>", "<title>one</title> <title>two</title>")
print(m)

#La expresión regular r"<title>((((?!<).)*))</title>" está diseñada para buscar contenido dentro de etiquetas <title>...</title>.
#La expresión regular coincide con la primera etiqueta <title>, que contiene "one", sin consumir el texto de la siguiente etiqueta <title>.
#La afirmación de anticipación negativa ((?!<)) asegura que la búsqueda se detenga al llegar a la próxima apertura de una etiqueta, lo que permite capturar solo el contenido de la primera etiqueta <title>.
```

```python
#Cuando se imprime m, se obtiene un objeto de coincidencia que incluye la cadena completa <title>one</title>.
#Al acceder al grupo 1 con m.group(1), se obtiene solamente el contenido "one".

print(m.group(1))  # 'one'
```

```python
#La función de sustitución Usando Grupos en expresiones regulares permite reemplazar partes de una cadena que coinciden con un patrón específico. 
# Utilizando grupos en la expresión regular se puede hacer referencia a partes de la coincidencia y usarlas en la cadena de reemplazo.

result = re.sub(r"(\w+)", "bla (before \g<1>)", "A phrase")
print(result)

#La función re.sub busca cada coincidencia en la cadena "A phrase".
#Para la primera coincidencia "A", se sustituye por "bla (before A)".
#Para la segunda coincidencia "phrase", se sustituye por "bla (before phrase)".
```
