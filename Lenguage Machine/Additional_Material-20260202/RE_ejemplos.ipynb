```python
#Matching a specific string: r"hello" matches the exact string 'hello'.
# Example 1: Matching a Specific String
def match_specific_string():
    """Matches 'hello' and returns True if found."""
    pattern = r"hello"
    string = "hello world"
    match = re.search(pattern, string)
    return bool(match)

print("Example 1:", match_specific_string())
```

```python
def match_specific_string():
    """Matches 'hello' and prints the match if found."""
    pattern = r"hello"
    string = "hello world"
    match = re.search(pattern, string)
    
    if match:
        # Print the matched string 'hello'
        print(f"Matched string: {match.group()}")
    else:
        print("No match found for 'hello'.")

# Call the function to see the output
print("Example 1:")
match_specific_string()
```

```python
#Matching a character sequence: r"abc" matches the characters 'a', 'b', and 'c' in that order.
# Example 2: Matching a Character Sequence
def match_character_sequence():
    """Matches 'abc' and returns True if found."""
    pattern = r"abc"
    string = "hello abc world"
    match = re.search(pattern, string)
    return bool(match)

print("Example 2:", match_character_sequence())
```

```python
def match_character_sequence():
    """Matches 'abc' and prints the match if found."""
    pattern = r"abc"
    string = "hello abc world"
    match = re.search(pattern, string)
    
    if match:
        # Print the matched sequence 'abc'
        print(f"Matched sequence: {match.group()}")
    else:
        print("No match found for 'abc'.")

# Call the function to see the output
print("Example 2:")
match_character_sequence()
```

```python
#Matching a character set: r"[ab]" matches either the character 'a' or 'b'.
# Example 3: Matching a Character Set
def match_character_set():
    """Matches 'a' or 'b' and returns True if found."""
    pattern = r"[ab]"
    string = "hello b world"
    match = re.search(pattern, string)
    return bool(match)

print("Example 3:", match_character_set())
```

```python
def match_character_set():
    """Matches 'a' or 'b' and prints the match if found."""
    pattern = r"[ab]"
    string = "hello b world"
    match = re.search(pattern, string)
    
    if match:
        # Print the matched 'a' or 'b'
        print(f"Matched character: {match.group()}")
    else:
        print("No match found for 'a' or 'b'.")

# Call the function to see the output
print("Example 3:")
match_character_set()
```

```python
#Matching a character range: r"[a-z]" matches any letter from 'a' to 'z'.
# Example 4: Matching a Character Range
def match_character_range():
    """Matches any letter from 'a' to 'z' and returns True if found."""
    pattern = r"[a-z]"
    string = "hello World"
    match = re.search(pattern, string)
    return bool(match)

print("Example 4:", match_character_range())
```

```python
def match_character_range():
    """Matches any letter from 'a' to 'z' and prints the match if found."""
    pattern = r"[a-z]"
    string = "hello World"
    match = re.search(pattern, string)
    
    if match:
        # Print the matched letter from 'a' to 'z'
        print(f"Matched letter: {match.group()}")
    else:
        print("No match found for any letter in the range 'a' to 'z'.")

# Call the function to see the output
print("Example 4:")
match_character_range()
```

```python
#Matching a quantifier (Zero or More): r"a*" matches the character 'a' zero or more times.
# Example 5: Matching a Quantifier (Zero or More)
def match_quantifier_zero_or_more():
    """Matches 'a' zero or more times and returns True if found."""
    pattern = r"a*"
    string = "hello world"
    match = re.search(pattern, string)
    return bool(match)

print("Example 5:", match_quantifier_zero_or_more())
```

```python
def match_quantifier_zero_or_more():
    """Matches 'a' zero or more times and prints the match if found."""
    pattern = r"a*"
    string = "hello world"
    match = re.search(pattern, string)
    
    if match:
        # Print the matched sequence of 'a' zero or more times
        print(f"Matched sequence: '{match.group()}'")
    else:
        print("No match found for 'a' zero or more times.")

# Call the function to see the output
print("Example 5:")
match_quantifier_zero_or_more()
```

```python
#Matching a quantifier (One or More): r"a+" matches the character 'a' one or more times.
# Example 6: Matching a Quantifier (One or More)
def match_quantifier_one_or_more():
    """Matches 'a' one or more times and returns True if found."""
    pattern = r"a+"
    string = "hello world aaabbb"
    match = re.search(pattern, string)
    return bool(match)

print("Example 6:", match_quantifier_one_or_more())
```

```python
#Matching a quantifier (Exactly N): r"a{3}" matches the character 'a' exactly three times.
# Example 7: Matching a Quantifier (Exactly N)
def match_quantifier_exactly_n():
    """Matches 'a' exactly three times and returns True if found."""
    pattern = r"a{3}"
    string = "hello world aaa"
    match = re.search(pattern, string)
    return bool(match)

print("Example 7:", match_quantifier_exactly_n())
```

```python
#Matching a word boundary: \b matches a word boundary, and r"\babc\b" matches the whole word 'abc'.
# Example 8: Matching a Word Boundary
def match_word_boundary():
    """Matches 'abc' as a whole word and returns True if found."""
    pattern = r"\babc\b"
    string = "hello abc world abcdef"
    match = re.search(pattern, string)
    return bool(match)

print("Example 8:", match_word_boundary())
```

```python
def match_word_boundary():
    """Matches 'abc' as a whole word and prints the match if found."""
    pattern = r"\babc\b"
    string = "hello abc world abcdef"
    match = re.search(pattern, string)
    
    if match:
        # Print the matched word
        print(f"Matched word: {match.group()}")
    else:
        print("No match found for the whole word 'abc'.")

# Call the function to see the output
print("Example 8:")
match_word_boundary()
```

```python
#Matching a non-word character: \W matches any character that is not a letter or digit, and r"\W" matches any such character.
# Example 9: Matching a Non-Word Character
def match_non_word_character():
    """Matches any character that is not a letter or digit and returns True if found."""
    pattern = r"\W"
    string = "hello world! @#$"
    match = re.search(pattern, string)
    return bool(match)

print("Example 9:", match_non_word_character())
```

```python
def match_non_word_character():
    """Matches any character that is not a letter or digit and prints the first match."""
    pattern = r"\W"
    string = "hello world! @#$"
    match = re.search(pattern, string)
    
    if match:
        # Print the matched non-word character
        print(f"Matched non-word character: {match.group()}")
    else:
        print("No non-word character found.")

# Call the function to see the output
print("Example 9:")
match_non_word_character()
```

```python
#Matching a word character: \w matches any character that is a letter or digit, and r"\w" matches any such character.
# Example 10: Matching a Word Character
def match_word_character():
    """Matches any character that is a letter or digit and returns True if found."""
    pattern = r"\w"
    string = "hello world123"
    match = re.search(pattern, string)
    return bool(match)

print("Example 10:", match_word_character())
```
