```python
import tensorflow as tf
import numpy as np

# Create from NumPy array
a = np.array([1, 2, 3], dtype=np.int32)
t_a = tf.convert_to_tensor(a)

# Create from Python list
b = [4, 5, 6]
t_b = tf.convert_to_tensor(b)

print(t_a)
print(t_b)
```

```python
tf.random.set_seed(1)
t1 = tf.random.uniform(shape=(5, 2), minval=-1.0, maxval=1.0)
t2 = tf.random.normal(shape=(5, 2), mean=0.0, stddev=1.0)

# Element-wise multiplication
t3 = tf.multiply(t1, t2).numpy()

# Calculate Mean
t4 = tf.math.reduce_mean(t1, axis=0)

print(t3)
print(t4)
```

```python
A = tf.ones((3,))
B = tf.zeros((3,))

# Concat: Joins along an existing axis (extends length)
C = tf.concat([A, B], axis=0) 
# Output: [1. 1. 1. 0. 0. 0.]

# Stack: Joins along a new axis (creates a matrix)
S = tf.stack([A, B], axis=1)
# Output: [[1. 0.], [1. 0.], [1. 0.]]
```

```python
a = [1.2, 3.4, 7.5, 4.1, 5.0, 1.0]
ds = tf.data.Dataset.from_tensor_slices(a)

# Batching data (e.g., batch size 3)
ds_batch = ds.batch(3)

for i, elem in enumerate(ds_batch, 100):
    print('batch {}'.format(i), elem.numpy())
```

```python
# Create data (X) and labels (y)
X = tf.random.uniform([4, 3], dtype=tf.float32)
t_y = tf.range(4)

# Create datasets
ds_x = tf.data.Dataset.from_tensor_slices(X)
ds_y = tf.data.Dataset.from_tensor_slices(t_y)

# Zip them together
ds_joint = tf.data.Dataset.zip((ds_x, ds_y))

for example in ds_joint:
    print('x:', example[0].numpy(), 'y:', example[1].numpy())
```
