```python
from sklearn.metrics import (confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,cohen_kappa_score)

y_true = [1]*50 + [0]*950      # 50 fraudes, 950 no fraudes
y_pred = [1]*40 + [0]*10 + [1]*90 + [0]*860

# Matriz de confusión
cm = confusion_matrix(y_true, y_pred)
print(cm)

# Métricas
print("Accuracy:", accuracy_score(y_true, y_pred))
print("Error rate:", 1 - accuracy_score(y_true, y_pred))
print("Sensitivity (Recall):", recall_score(y_true, y_pred))
print("Specificity:", cm[0,0] / (cm[0,0] + cm[0,1]))
print("FP rate:", cm[0,1] / (cm[0,0] + cm[0,1]))
print("Precision:", precision_score(y_true, y_pred))
print("F1:", f1_score(y_true, y_pred))
print("Kappa:", cohen_kappa_score(y_true, y_pred))
```

Sklearn siempre devuelve la confusion matrix de esta manera:
[[VN FP]
 [FN VP]]
