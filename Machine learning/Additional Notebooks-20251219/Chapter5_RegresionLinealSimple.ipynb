## Regresión Lineal Simple
Construir un modelo de regresión lineal simple para predecir la calificación de un estudiante en un examen basado en el número de horas que estudia.

-Exploración de datos \
-Preparación de datos \
-Construcción del modelo de regresión lineal \
-Evaluación del modelo

```python
#Libraries import
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import dataset
path = pd.read_csv("student_scores.csv")
df = path
```

```python
#Construir un modelo de regresión lineal simple para predecir la calificación de un estudiante en un examen basado en el número de horas que estudia.
#Exploración de datos
print(df.head())
```

```python
sns.scatterplot(x='Hours', y='Scores', data=df)
plt.title('Horas de estudio vs Calificaciones')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.show()
```

```python
# Definimos variables X e Y
X = df['Hours']
y = df['Scores']

# Usamos numpy polyfit para una regresión de grado 1 (lineal)
# Esto calcula m (pendiente) y b (intercepto) minimizando el error cuadrático
m, b = np.polyfit(X, y, 1)

print(f"Pendiente (m): {m}")
print(f"Intercepto (b): {b}")
```

```python
# Ecuación de la recta: y = mx + b
plt.figure(figsize=(8,5))
sns.scatterplot(x='Hours', y='Scores', data=df, label='Datos Reales')

# Trazamos la línea roja usando la ecuación calculada
plt.plot(X, m*X + b, color='red', label='Línea de Regresión')

plt.title('Regresión Lineal (Solo NumPy)')
plt.legend()
plt.show()
```
