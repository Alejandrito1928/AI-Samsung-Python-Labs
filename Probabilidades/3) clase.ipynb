```python
import matplotlib.pyplot as plt
from scipy.stats import binom
import numpy as np

# 1. Configuración de datos (Basado en la imagen)
n = 10
p = 0.7
x = np.arange(0, n + 1)           # Valores del eje X (0 al 10)
probabilidades = binom.pmf(x, n, p) # Calculamos la altura de cada punto


fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(x, probabilidades, color='#1f449c', s=100, zorder=3) 


ax.set_facecolor('#f4f4f4')      
fig.patch.set_facecolor('white')  

ax.set_title('Bernoulli random variable and probability distribution', loc='left', pad=20)
ax.text(5, 0.30, r'$n=10, p=0.7$', fontsize=18, ha='center', color='#001133')


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

# Límites y Ticks (Etiquetas del eje)
ax.set_xlim(-0.5, 10.5)
ax.set_ylim(0, 0.35)
ax.set_xticks(np.arange(0, 11, 2)) # Mostrar solo 0, 2, 4...

plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import binom
import numpy as np

# --- PASO 1: LOS DATOS (Matemática) ---
n = 10    # Número de intentos
p = 0.7   # Probabilidad de éxito
x = np.arange(0, n + 1)           # Eje X: Números del 0 al 10
y = binom.pmf(x, n, p)            # Eje Y: Probabilidad exacta de cada uno

# --- PASO 2: EL GRÁFICO (Pintar) ---
plt.figure(figsize=(8, 5))  # Crea un lienzo de 8x5 pulgadas

# Usamos 'scatter' porque queremos PUNTOS, no una línea continua
plt.scatter(x, y, color='blue', s=100)  # s=100 es el tamaño del punto

# --- PASO 3: ETIQUETAS (Para que se entienda) ---
plt.title(f'Distribución Binomial (n={n}, p={p})') # Título arriba
plt.xlabel('Número de éxitos')  # Etiqueta abajo
plt.ylabel('Probabilidad')      # Etiqueta izquierda

# --- PASO 4: MOSTRAR ---
plt.grid(True, alpha=0.3)       # Cuadrícula suave para guiar el ojo
plt.show()                      # ¡Foto!
```
