```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
```

```python
#La altura de los estudiantes de una universidad se distrubye normalmente con media 170 cm y desviación estándar 10 cm.
#¿Cuál es la probabilidad de que un estudiante seleccionado al azar mida entre 160 cm y 180 cm?
media = 170
anchura = 10 
rango_min = 160
rango_max = 180 

prob_hasta_180 = st.norm.cdf(rango_max, media, anchura)

prob_hasta_160 = st.norm.cdf(rango_min, media, anchura)

prob_final = prob_hasta_180 - prob_hasta_160

print(f"--- Resultados ---")
print(f"Probabilidad acumulada hasta 180cm: {prob_hasta_180:.4f}")
print(f"Probabilidad acumulada hasta 160cm: {prob_hasta_160:.4f}")
print(f"✅ Probabilidad ENTRE 160 y 180: {prob_final:.4f} ({prob_final*100:.2f}%)")

z_min = (rango_min - media) / anchura
z_max = (rango_max - media) / anchura
print(f"Check Z-Scores: De {z_min} a {z_max} sigmas.")
```

```python
x = np.linspace(media - 4*anchura, media + 4*anchura, 1000)
y = st.norm.pdf(x, media, anchura)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='black', linewidth=2)

mask = (x >= rango_min) & (x <= rango_max)
plt.fill_between(x[mask], y[mask], color='skyblue', alpha=0.5)

plt.axvline(media, color='red', linestyle='--')
plt.axvline(rango_min, color='green', linestyle=':')
plt.axvline(rango_max, color='green', linestyle=':')

plt.title('Distribución Normal de Alturas')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad')
plt.show()
```

```python
#La puntuaacion de un examen se distrivuye nomralmente con media 50 y desviacion estandar 5. 
# ¿ Cual es la probbilidad de que un estudiante saque entre 45 y 55
mu = 50        # Media
sigma = 5      # Desviación
min_val = 45   # Límite inferior
max_val = 55   # Límite superior


prob = st.norm.cdf(max_val, mu, sigma) - st.norm.cdf(min_val, mu, sigma)

print(f"Probabilidad: {prob:.4f}")
```

```python

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = st.norm.pdf(x, mu, sigma)

plt.figure(figsize=(10, 6))

plt.plot(x, y, color='darkblue', linewidth=2, label='Distribución Normal')

mask = (x >= min_val) & (x <= max_val)
plt.fill_between(x[mask], y[mask], color='orange', alpha=0.5, 
                 label=f'Probabilidad: {prob:.2%}')


plt.axvline(mu, color='red', linestyle='--', label='Media (50)')
plt.axvline(min_val, color='green', linestyle=':', label='Min (45)')
plt.axvline(max_val, color='green', linestyle=':', label='Max (55)')

plt.title('Distribución de Notas del Examen')
plt.xlabel('Puntuación')
plt.ylabel('Densidad')
plt.legend()
plt.grid(alpha=0.3)

plt.show()
```
