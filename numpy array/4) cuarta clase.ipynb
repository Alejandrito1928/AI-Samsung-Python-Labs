```python
import pandas as pd

# 1. CREAMOS LOS DATOS
# Tabla A: Empleados (ID y Nombre)
df_empleados = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Nombre': ['Ana', 'Beto', 'Carla', 'David'],
    'Depto': ['Ventas', 'IT', 'Ventas', 'RRHH']
})

# Tabla B: Sueldos (ID y Salario) - Fíjate que falta el ID 4 y sobra el ID 5
df_sueldos = pd.DataFrame({
    'ID': [1, 2, 3, 5],
    'Salario': [2500, 3000, 2800, 4000]
})


```

```python
print("--- Tabla Empleados ---")
print(df_empleados)
print("\n--- Tabla Sueldos ---")
print(df_sueldos)
```

```python

# 2. INNER JOIN (Solo los que están en ambas listas)
# David (4) desaparece porque no tiene sueldo. El 5 desaparece porque no es empleado.
print("\n--- 1. INNER JOIN (Intersección) ---")
print(pd.merge(df_empleados, df_sueldos, on='ID', how='inner'))
```

```python

# 3. LEFT JOIN (Prioridad Empleados)
# David (4) se queda, pero su sueldo será NaN (desconocido).
print("\n--- 2. LEFT JOIN (Manda Empleados) ---")
print(pd.merge(df_empleados, df_sueldos, on='ID', how='left'))
```

```python
# 4. OUTER JOIN (Todos dentro)
# Entran todos: David (sin sueldo) y el ID 5 (sueldo sin nombre).
print("\n--- 3. OUTER JOIN (Todos) ---")
print(pd.merge(df_empleados, df_sueldos, on='ID', how='outer'))
```

```python
import pandas as pd

# 1. CREAMOS LOS DATOS
# Tabla A: Empleados (ID y Nombre)
df_sueldos = pd.DataFrame({
    'ID': [1, 2, 3, 5],
    'Salario': [2500, 3000, 2800, 4000]
})
#ordenar sort_values 
df = df_sueldos.sort_values(by='ID')
print("Sorted array based on keys:")
print(df.head())
#descendente
df_desc = df_sueldos.sort_values(by='ID', ascending=False)
print("Sorted array in descending order based on keys:")
print(df_desc.head())
#ordenar por clase y sueldo
df_multi = df_sueldos.sort_values(by=['Salario', 'ID'])
print("Sorted array based on Salario and ID:")
print(df_multi.head())
```

```python
import pandas as pd

# --- CREAR DATAFRAME A ---
data_a = {
    'Name': ['Harry Potter', 'David Baker', 'John Smith', 'Juan Martinez', 'Jane Connor'],
    'Gender': ['Male', 'Male', 'Male', 'Male', 'Female'],
    'Age': [23, 31, 22, 36, 30]
}

df_A = pd.DataFrame(data_a)

# --- CREAR DATAFRAME B ---
data_b = {
    'Name': ['John Smith', 'Alex Du Bois', 'Joanne Rowling', 'Jane Connor'],
    'Position': ['Intern', 'Team Lead', 'Manager', 'Manager'],
    'Wage': [25000, 75000, 90000, 70000]
}

df_B = pd.DataFrame(data_b)

# --- VERIFICACIÓN ---
print("--- DataFrame A ---")
print(df_A)
print("\n--- DataFrame B ---")
print(df_B)
```

```python
# --- 1. INNER JOIN (Intersección) ---
# Se queda SOLO con los que están en ambas listas (John Smith y Jane Connor).
print("\n--- 1. INNER JOIN (how='inner') ---")
# Opción corta (por defecto es inner)
inner_1 = pd.merge(df_A, df_B, on='Name') 
# Opción explícita (como en la imagen)
inner_2 = pd.merge(df_A, df_B, on='Name', how='inner') 
print(inner_2)

```

```python

# --- 2. LEFT JOIN (Prioridad Izquierda) ---
# Mantiene a TODOS los de df_A (Harry, David, Juan). 
# Si no encuentra coincidencia en B, pone NaN.
print("\n--- 2. LEFT JOIN (how='left') ---")
left_join = pd.merge(df_A, df_B, on='Name', how='left')
print(left_join)
```

```python


# --- 3. RIGHT JOIN (Prioridad Derecha) ---
# Mantiene a TODOS los de df_B (Alex Du Bois, Joanne Rowling).
# Si no están en A, sus datos de 'Gender' y 'Age' serán NaN.
print("\n--- 3. RIGHT JOIN (how='right') ---")
right_join = pd.merge(df_A, df_B, on='Name', how='right')
print(right_join)
```

```python


# --- 4. FULL OUTER JOIN (Unión Total) ---
# Nadie se queda fuera. Junta todo.
# Verás muchos NaNs porque rellena los huecos de ambos lados.
print("\n--- 4. FULL OUTER JOIN (how='outer') ---")
outer_join = pd.merge(df_A, df_B, on='Name', how='outer')
print(outer_join)
```

```python
#ahora concatenaremos
import pandas as pd

# --- CREAR DATAFRAME A ---
data_a = {
    'Name': ['Harry Potter', 'David Baker', 'John Smith', 'Juan Martinez', 'Jane Connor'],
    'Gender': ['Male', 'Male', 'Male', 'Male', 'Female'],
    'Age': [23, 31, 22, 36, 30]
}

df_A = pd.DataFrame(data_a)

# --- CREAR DATAFRAME B ---
data_b = {
    'Name': ['John Smith', 'Alex Du Bois', 'Joanne Rowling', 'Jane Connor'],
    'Position': ['Intern', 'Team Lead', 'Manager', 'Manager'],
    'Wage': [25000, 75000, 90000, 70000]
}

df_B = pd.DataFrame(data_b)

# --- VERIFICACIÓN ---
print("--- DataFrame A ---")
print(df_A)
print("\n--- DataFrame B ---")
print(df_B)
# --- CONCATENAR VERTICALMENTE (FILAS) ---
print("pd.concat([df_A, df_B], sort=True)")
print(pd.concat([df_A, df_B], sort=True))
print("pd.concat([df_A, df_B], axis = 1, sort = True)")
print(pd.concat([df_A, df_B], axis = 1, sort = True))
```
