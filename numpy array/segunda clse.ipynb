```python
#crea un array numpy de 5000 x 5000 25 millones de elementos llenos de 0 
#realizar el ejericio de dos maneros 
# usando un buble for anidado para llena runa lista de listas y luego convertirla a un array de numpy 
#usando unfcion np.zeros
import numpy as np
import time

array_size = 5000

print(f"--- INICIO: Creando matriz de {array_size}x{array_size} ---")

# ---------------------------------------------------------
# FORMA 1: Bucle For Anidado (La forma "Manual")
# ---------------------------------------------------------
start_manual = time.time()

lista_de_listas = []
for i in range(array_size):
    fila = []
    for j in range(array_size):
        fila.append(0)     # Agregamos el 0 uno por uno
    lista_de_listas.append(fila)

# Convertimos esa lista gigante a NumPy (esto también tarda)
array1 = np.array(lista_de_listas)

end_manual = time.time()
tiempo_manual = end_manual - start_manual
print(f"Tiempo Bucle For: {tiempo_manual:.5f} segundos")


# ---------------------------------------------------------
# FORMA 2: np.zeros (La forma NumPy)
# ---------------------------------------------------------
start_numpy = time.time()

# np.zeros reserva la memoria de golpe. No itera en Python.
array2 = np.zeros((array_size, array_size))

end_numpy = time.time()
tiempo_numpy = end_numpy - start_numpy
print(f"Tiempo np.zeros:  {tiempo_numpy:.5f} segundos")


# ---------------------------------------------------------
# CONCLUSIÓN
# ---------------------------------------------------------
if tiempo_numpy > 0:
    velocidad = tiempo_manual / tiempo_numpy
    print(f"\nNumPy fue {velocidad:.0f} veces más rápido.")
else:
    print("\nNumPy fue instantáneo (0.00 segundos).")

# Verificación rápida de que son iguales en forma
print(f"\nForma Array 1: {array1.shape}")
print(f"Forma Array 2: {array2.shape}")

```

```python
import numpy as np

# 0. Creamos el array base
arr = np.arange(10)
print(f"Inicio: {arr}")

# --- PARTE 1: VISTA (Modifica el original) ---
vista = arr[2:6]        # Esto es solo una ventana, no un array nuevo
vista[0] = 999          # Cambiamos el primer valor de la vista

print("\n--- VISTA ---")
print(f"Vista: {vista}")
print(f"Arr:   {arr}")  # ¡OJO! El 999 apareció aquí también

# --- PARTE 2: COPIA (Independiente) ---
copia = arr[2:6].copy() # .copy() crea un array nuevo en memoria
copia[0] = 111          # Cambiamos la copia

print("\n--- COPIA ---")
print(f"Copia: {copia}")
print(f"Arr:   {arr}")  # El original NO cambia (sigue con el 999 de antes)
```
