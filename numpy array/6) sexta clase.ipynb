```python
dat = sns.load_dataset("mpg")
dat.shape
dat.head(5)
```

```python
dat["mpg"]
```

```python
sns.displot(dat.mpg, kde=False, rug=True, bins=50, color="Red")
#kdf es para que salga la curva de distribución 
plt.show()
```

```python
sns.displot(dat.mpg, kde=True, rug=False, bins=50, color="Blue")
#kdf es para que salga la curva de distribución 
plt.show()
```

```python
sns.displot(dat.mpg, kde=False, rug=True, bins=50, color="Blue")
#kdf es para que salga la curva de distribución 
plt.show()
```

```python
sns.kdeplot(dat.mpg, color="blue")
plt.show()
```

```python
sns.jointplot(x="weight", y="mpg", data=dat, kind="scatter", color="purple")
plt.show()
```

```python
sns.jointplot(x="weight", y="mpg", data=dat, kind="reg", color="orange")
plt.show()
```

```python
sns.lmplot(data= dat, x = "weight", y= "mpg")
plt.show()
```

```python
sns.violinplot(x= "cylinders", y="horsepower", data= dat, palette="coolwarm")
plt.show()
```

```python
g = sns.FacetGrid(dat, col="cylinders")
g.map(sns.scatterplot, "weight", "mpg")
plt.show()
```

```python
#heat.map
dat = sns.load_dataset("mpg")
dat.drop(columns= ["origin", "name"], inplace=True)
x = dat.corr()
x
```
