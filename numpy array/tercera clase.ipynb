```python
import numpy as np

cubo = np.arange(6).reshape(2, 3, 1)

print("--- ORIGINAL (2, 3, 1) ---")
print(cubo)
print(cubo.shape)


print("CASO A: Transpose (.T) - INVIERTE TODO")

# El eje 0 pasa al final, el eje 2 pasa al principio.
# De (2, 3, 1) pasa a (1, 3, 2)
transpuesta = cubo.T

print("\n--- TRANSPOSE .T (Invierte todo el orden) ---")
print(transpuesta)
print(f"Forma nueva: {transpuesta.shape}")  # (1, 3, 2)


# ---------------------------------------------------------
# CASO B: Swapaxes (0, 1) - SOLO CAMBIA EL INICIO
# ---------------------------------------------------------
# Le decimos: "Intercambia solo el Bloque (0) con la Fila (1). Deja la Columna (2) quieta".
# De (2, 3, 1) pasa a (3, 2, 1)
intercambio = np.swapaxes(cubo, 0, 1)

print("\n--- SWAPAXES 0,1 (Solo mueve los dos primeros) ---")
print(intercambio)
print(f"Forma nueva: {intercambio.shape}")  # (3, 2, 1)
```

```python
#para que sirve swapaxes y transpose no sirve mucho 
import numpy as np

# Simulemos una imagen: (Alto=2, Ancho=2, Canales=3)
# Imagina que son los valores RGB de 4 píxeles
imagen_original = np.zeros((2, 2, 3))
print(f"Forma Original (Para mostrar en pantalla): {imagen_original.shape}")

# --- INTENTO 1: Usar .T (El Error) ---
# Esto invierte (3, 2, 2), PERO mezcla filas con columnas
intento_t = imagen_original.T
print(f"Con .T (Gira la imagen):                   {intento_t.shape}")

# --- INTENTO 2: Usar moveaxis/swapaxes (Lo Correcto para IA) ---
# Queremos mover el eje 2 (Canales) al eje 0, sin rotar la foto.
imagen_pytorch = np.moveaxis(imagen_original, 2, 0) # Mueve el eje 2 a la posición 0
print(f"Para PyTorch (Canales primero, sin girar): {imagen_pytorch.shape}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. CREAR UNA IMAGEN RECTANGULAR (No Cuadrada) ---
# Usamos 50 de alto x 100 de ancho para notar si se gira.
# Formato inicial: (Alto, Ancho, Canales) -> (50, 100, 3)
imagen = np.zeros((50, 100, 3), dtype=np.uint8)

# Pintamos: Mitad Izquierda ROJA, Mitad Derecha AZUL
imagen[:, :50, 0] = 255   # Rojo a la izquierda
imagen[:, 50:, 2] = 255   # Azul a la derecha

print(f"1. Forma Original: {imagen.shape} (Rectángulo ancho)")

# Mostramos la original
plt.figure(figsize=(10,3))
plt.subplot(1, 3, 1)
plt.title("1. Original (Ancha)")
plt.imshow(imagen)


# --- 2. EL ERROR: USAR .T (Transpose) ---
# Intentas preparar los datos para la IA (Canales primero) usando .T
# De (50, 100, 3) pasa a (3, 100, 50).
imagen_rota = imagen.T

print(f"2. Forma con .T:   {imagen_rota.shape} (Canales, Ancho, Alto) -> ¡Mezcló dimensiones!")

# PARA VER EL ERROR:
# No usamos .T de nuevo (eso lo arreglaría).
# Movemos el eje de canales (que quedó en el 0) al final para que matplotlib lo pinte.
# Esto nos mostrará cómo quedó la estructura espacial.
visualizacion_error = np.moveaxis(imagen_rota, 0, -1)

plt.subplot(1, 3, 2)
plt.title("2. Error con .T (Girada)")
plt.imshow(visualizacion_error) 
# NOTA: Verás que la imagen se puso de pie (50 de ancho x 100 de alto). ¡Se rompió!


# --- 3. LA FORMA CORRECTA: MOVEAXIS ---
# Movemos el eje de color (2) al principio (0), respetando Alto y Ancho.
# De (50, 100, 3) pasa a (3, 50, 100)
imagen_correcta_ia = np.moveaxis(imagen, -1, 0)

print(f"3. Forma IA:       {imagen_correcta_ia.shape} (Canales, Alto, Ancho) -> ¡Correcto!")

# Para visualizarla, revertimos el movimiento suavemente
visualizacion_correcta = np.moveaxis(imagen_correcta_ia, 0, -1)

plt.subplot(1, 3, 3)
plt.title("3. Correcta (Moveaxis)")
plt.imshow(visualizacion_correcta)
# NOTA: Sigue siendo un rectángulo ancho.

plt.show()
```
